```python
translation = {}
for sentence in to_translate:
    translation[(sentence)] = GoogleTranslator(source="sv", target="en").translate(sentence)
```

```python
to_learn = {k.capitalize(): v.capitalize() for k,v in translation.items()}
```

```python
my_model = genanki.Model(
    1607392319,
    'Simple Model',
    fields=[
        {'name': 'Front'},
        {'name': 'Back'},
    ],
    templates=[
        {
            'name': 'Card 1',
            'qfmt': '{{Front}}',  # what shows on front
            'afmt': '{{FrontSide}}<hr id="answer">{{Back}}',  # what shows on back
        },
    ])

# Create a deck
my_deck = genanki.Deck(
    2059200110,
    'Swedish Vocabulary BF1 - finall')
```

```python
for swedish, english in to_learn.items():
    # Add cards
    my_deck.add_note(genanki.Note(
        model=my_model,
        fields=[english, swedish]))
```

```python
# Export to .apkg
genanki.Package(my_deck).write_to_file('swedish_vocab.apkg')
```
